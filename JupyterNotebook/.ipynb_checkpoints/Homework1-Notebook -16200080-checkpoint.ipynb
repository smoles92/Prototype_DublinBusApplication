## Exercises

This homework focuses on data exploration and preparation for a particular problem and dataset.
The problem and data come from the Amazon online shopping platform. Several sellers can sell the same product on Amazon. Based on the data provided by the seller to Amazon (seller reputation, product price, shipping details, etc) Amazon ranks seller offers from best to worst for a given product. This ranking is mostly influenced by the product price offer of the seller, but it can also be influenced by other features. The goal is to explore and transform the data to try and understand which features are most indicative of a seller being ranked first by Amazon for a product. When the seller is ranked first for a product we say that the seller is the 'winner' among all the offers, because their offer is shown first when a user searches for a product on Amazon, which increases their chances of selling the product.

We have a CSV file describing offers by given sellers for given products and a column which records whether an offer was a winner or not.

#### 3. Outliers
The observations noted between the mean and the median on continuous features, the exponential histograms and the box plots provide a strong indication of the presence of outliers. 

##### Shipping Times.

```python
df.groupby(['ShippingTime_minHours', 'SellerId']).count()
```

```python
df.groupby(['ShippingTime_maxHours', 'SellerId']).count()
```

In total, 235 fields are missing shipping time information. Of the 235, one seller is missing 224 fields in shipping times. The seller is "Fulfilled by Amazon" which means that they are a third-party seller, but ships from an Amazon Fulfilment Centre to the customer. All Amazon.com standard shipping rates and policies apply to these items including free shipping on qualifying orders and Prime shipping. The seller affects 2% of the data through this incomplete information. Since, standard shipping policies apply here anyway, carrying out imputation using the median is a valid approach to use here in order to create more accurate model. The median for min hours is 24, and max is 48. The 11 other fields incorrectly marked as zero for min and maximum shipping times will also be altered.

```python
df['ShippingTime_minHours'] = df['ShippingTime_minHours'].replace([0], [24])
```

```python
df['ShippingTime_maxHours'] = df['ShippingTime_maxHours'].replace([0], [48])
```

Since the range for ‘ShippingTime_minHours’ and ‘ShippingTime_maxHours' is still large, the outliers persist which makes it difficult to plot. Therfore, instead of using hours, the data will be converted to days. In addition, since both feature don’t provide much value in isolation, they can be concatenated to easily compare the relationship between the two. In doing this, the cardinality will be reduced substantially and the feature will become categorical. 

```python
# First, convert data from hours to days.
df['ShippingTime_minDays'] = df['ShippingTime_minHours']//24
df['ShippingTime_maxDays'] = df['ShippingTime_maxHours']//24
```

```python
# Create new column which concatenates both features.
df['ShippingTime'] = df['ShippingTime_minDays'].astype(str) + "-" + df['ShippingTime_maxDays'].astype(str) + " days"
```

```python
# Drop other columns no longer of use.
df = df.drop(['ShippingTime_maxDays','ShippingTime_minDays','ShippingTime_maxHours', 'ShippingTime_minHours'], 1)
```

```python
# Replace values that have 0-0 days with just a single 0 days.
df['ShippingTime'] = df['ShippingTime'].replace(['0-0 days'], ['0 days'])
```

```python
# Mark item now as a categorical feature.
df['ShippingTime'] = df['ShippingTime'].astype('category')
```

```python
df['ShippingTime'].value_counts().plot(kind='bar', figsize=(12,5), color="black", alpha=0.3,title ="Ships From Country")
```

First key takeaway is that only featured merchants can win. Since this amounts to 7543, this rules out a possible 1796 straight away because they are not featured merchants. However, being 'fulfilled by amazon' doesn't guarantee that you win all the time, this will require further information.

```python
sns.set(style="white")

# Select columns containing continuous data
continuous_columns = clean_df[['ListingPrice','SellerFeedbackCount','SellerFeedbackRating','ShippingPrice', 'TotalPrice']].columns
corr = clean_df[continuous_columns].corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=1, vmin=-1,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)

```

```python
# List unique values for Shipping Times
ShippingTime = pd.unique(clean_df.ShippingTime.ravel())

# Create new Column and fill with Shipping Time %
clean_df['percent'] = 0

# Iterate through the unique values in ShippingTimes and for each value count the amount of Winning Sellers
# Find the indexes of each row in a particular ShippingTime and for each of these row insert count * 100 in the percent column
for c in ShippingTime:
    count = 1 / clean_df[clean_df.ShippingTime == c].count()['IsWinner']
    index_list = clean_df[clean_df['ShippingTime'] == c].index.tolist()
    for i in index_list:
        clean_df.loc[i, 'percent'] = count * 100
        
# Group dataframe by ShippingTimes and Winning Seller and sum
group = clean_df[['percent','ShippingTime','IsWinner']].groupby(['ShippingTime','IsWinner']).sum()

# Plot values of group in a stacked bar chart
my_plot = group.unstack().plot(kind='bar', stacked=True, title="Winning according to Shipping Times", figsize=(15,7))

# Define label colours and text and add to the plot
red_patch = mpatches.Patch(color='green', label='Winner')
blue_patch = mpatches.Patch(color='blue', label='Loser')
my_plot.legend(handles=[red_patch, blue_patch], frameon = True)

# Define x and y labels and min and max values for the y axis
my_plot.set_xlabel("Shipping Times")
my_plot.set_ylabel("Winning")
my_plot.set_ylim([0,100])
plt.tight_layout()
```

Overall from above, the winners only come from five bins. Typically, a winner will have a shipping time of less than 5 days. However, there are excepctions with three winners taking between 21-30 days. See below for more information.
