## <font color='#9d95d0'> 1.2. Business Objective </font>
*** Determine how Amazon chooses a seller of a product to be shown first when a user searchers for a product.***

For Amazon, several sellers can sell the same product on its web-based shopping platform. Based on the data provided by the seller to Amazon (seller reputation, product price, shipping details etc.) Amazon ranks seller offers from best to worst for a given product.

## <font color='#9d95d0'>1.3. Research Question </font>
*** What features are most indicative of a seller for an Amazon Product to be ranked first? ***

The goal of the assignment is to capture the relationship between descriptive features and the target feature 'IsWinner'. When the seller is ranked first for a product, they are declared a the 'winner' among all the offers. This means their offer is shown first when a user searches for a product on Amazon, which increases their chances of selling the product.

Sellers that are FulfilledByAmazon causing outliers in the dataset and skewing results. From our research, a seller being fulfilledbyamazon means that amazon themselves take care of the shipping on their behalf. The Seller is listed as ‘Fulfilled by Amazon’. This means that the merchant is a third-party seller, but ships from an Amazon Fulfilment Centre to the customer. Standard shipping rates and policies apply to these items including free shipping on qualifying orders (Amazon, 2017)

It is likely this is the reason amazon don't provide results for these. It's largely from one seller who causes 365 of these outliers. Considering a large number of IsFulFilledByAmazon are winners, these results negatively skew the data. Imputation to the mean for these values would be a better strategy for dealing with these outliers so they don't negatively sway the dataset. As such, all fields affected by this will be replaced with the mean value of the column. This affects: SellerFeedbackRating, SellerFeedbackCount, ShippingTime_minHours, ShippingTime_maxHours.

```python
avg1 = np.round(df['SellerFeedbackRating'].mean(), decimals=0)
avg2 = np.round(df['SellerFeedbackCount'].median(), decimals=0) # because two sellers distort the mean values by thousands.
avg3 = np.round(df['ShippingTime_minHours'].mean(), decimals=0)
avg4 = np.round(df['ShippingTime_maxHours'].mean(), decimals=0)

for i, row in df.iterrows():
        if df['IsFulfilledByAmazon'][i] == 1 and df['SellerFeedbackRating'][i] == 0 and df['SellerFeedbackCount'][i] == 0 and\
        df['ShippingTime_minHours'][i] == 0 and df['ShippingTime_maxHours'][i] == 0:
            df['SellerFeedbackRating'] = df['SellerFeedbackRating'].replace(0, avg1)
            df['SellerFeedbackCount'] = df['SellerFeedbackCount'].replace(0, avg2)
            df['ShippingTime_minHours'] = df['ShippingTime_minHours'].replace(0, avg3)
            df['ShippingTime_maxHours'] = df['ShippingTime_maxHours'].replace(0, avg4)
```

```python
# Created a color coding scheme for small, medium and large correlations to indicate the strength of association. All observations are then visually easy to extract for written observations (see report). See Pearsons guidelines: https://statistics.laerd.com/statistical-guides/pearson-correlation-coefficient-statistical-guide.php
# Styling Source: http://pandas.pydata.org/pandas-docs/stable/style.html

def highlight_small(val):
    color = 'green' if abs(val) > 0.1 and abs(val) <0.3 else ''
    return 'color: %s' % color

def highlight_medium(val):
    color = 'lightskyblue' if abs(val) > 0.3 and abs(val) <0.5 else ''
    return 'color: %s' % color

def highlight_large(val):
    color = 'red' if abs(val) > 0.5 else ''
    return 'color: %s' % color

# green is small association (0.1:0.3 or -0.1:-0.3)
# skyblue is medium association (0.3:0.5 or -0.3:-0.5)
# red is large association (0.5-1 or -0.5:-1)
```

## <font color='#9d95d0'>3.4. Normalisation of features</font>

Need to normalise three of the four selected features as they are viewed within the scope of the bidding process. When sellers compete for IsWinner, they compete over a product within a given timestamp. As such, we need to normalise TotalPrice, Shipping Times and Seller Feedback within this scope. 

```python
# In order to noramlise Shipping times, it needs to be converted to an integer. As there are only 6 possible combinations with 
# 1 being best and 6 being worst, we convert the categorical data to reflect this scale. 
df['Shipping_normed'] = df['ShippingTime'].map({'1-2 days' : 1, '2-3 days' : 2, '3-4 days' : 3, '4-5 days' : 4,\
                                                '6-10 days' : 5, 'More than 10 Days' : 6,})
```

```python
#Group the data frame by values needed
#Remove all instances of competitions where there are no winners or more than 1 winner
gb = df.groupby(['ProductId', 'TimeOfOfferChange'])

#Check how many there are
for _, group in gb:
    if sum(group['IsWinner']) !=1:
        df2 = df.drop(group.index)

#Description of the combinations
gb = df2.groupby(['ProductId', 'TimeOfOfferChange'])
gb.describe()
```

```python
def normalize(df):
    continuous_columns=['TotalPrice','SellerFeedbackRating','Shipping_normed']
    for columns in continuous_columns:
        df[columns]=(df[columns]-df[columns].min())/(df[columns].max()-(df[columns].min()))
    return df
```

```python
norm_df = gb.apply(normalize)
```

```python
print("Feature, Missing Values") 
norm_df.isnull().sum()
```

```python
# Replace 4 remaining Nan values
norm_df['Shipping_normed'].fillna((norm_df['Shipping_normed'].mean()), inplace=True)
```

```python
norm_df = norm_df.drop(['ListingPrice','SellerFeedbackCount', 'SellerId', 'ShippingTime', 'ShipsFromCountry','ShipsFromState', 'ShippingPrice'], 1)
```
