Created a color coding scheme for small, medium and large correlations to indicate the strength of association. All observations are then visually easy to extract for written observations (see report). See Pearsons guidelines: https://statistics.laerd.com/statistical-guides/pearson-correlation-coefficient-statistical-guide.php
Styling Source: http://pandas.pydata.org/pandas-docs/stable/style.html 

```python
def highlight_small(val):
    color = 'green' if abs(val) > 0.1 and abs(val) <0.3 else ''
    return 'color: %s' % color

def highlight_medium(val):
    color = 'lightskyblue' if abs(val) > 0.3 and abs(val) <0.5 else ''
    return 'color: %s' % color

def highlight_large(val):
    color = 'red' if abs(val) > 0.5 else ''
    return 'color: %s' % color
```

# Part 2: Predictive Modeling: Linear Regression

### (2.1) [5] Train a linear regression model to predict the target feature IsWinner, using the descriptive features selected in exercise (1). Evaluate the quality of the model on the training set.   

##### 2.1.1. Train the Model

```python
# For regression model to work with appropriately with target feature, the Yes/No columns needed to be changed to 1 or 0
# Otherwise, two features were plotted. 
df['IsWinner'] = df['IsWinner'].map(dict(yes=1, no=0))
```

In section one, the descriptive features selected were 1) continuous: a) Seller Feedback Rating and b) Total Price; 2) categorical: a) IsFulfilledByAmazon, b) ShippingTimes and c) Shipped From Country. As linear regression deals with continuous features, only those have been chosen to investigate. 

```python
#Import statsmodels package for training a linear regression model.
import statsmodels.formula.api as sm

# Look at the descriptive features selected in section 1. First: SellerFeedbackRating
lm1 = sm.ols(formula="IsWinner ~ SellerFeedbackRating", data=df).fit()
print(lm1.params)
```

```python
# Look at the descriptive features selected in section 1. Second: TotalPrice
lm2 = sm.ols(formula="IsWinner ~ TotalPrice", data=df).fit()
print(lm2.params)
```

#### 2.1.2. Observations on quality of descriptive feature for linear regression
- Only one of the selected sub-features is worth investigating further. For Total Price, it's interprative value is only worthy when viewed within the scope of the timestamp and for a the specific product. Analysing the Total price in itself won't provide any added value, especially since IsWinner is not distributed evenly across the dataset. More than 90% of the dataset are not winners to it's unlikely to pick up the effect of total price on IsWinner outside of the timestamp context. 
- For SellerFeedback Rating, it can be interpreted outside of the scope of the timestamp and product. As such, this is the example chosen for linear regression and not the both of them together. However, dealing with a categorical feature, there is a need to create thresholds to determine the effect of the coefficient on the target feature.

#### 2.1.3. Testing the model
Using the trained model to predict the target feature IsWinner, given the descriptive feature SellerFeedbackRating.

```python
X_new = pd.DataFrame({'SellerFeedbackRating': [99]})
X_new.head()
```

```python
# Predict the Winner for a new big with a SellerFeedbackRating, using the trained model. 
# The predicted value is 0.02 which is quite low. Let us check the interquartile range. 
lm1.predict(X_new)
```

```python
X_quartile = pd.DataFrame({'SellerFeedbackRating': [df.SellerFeedbackRating.min(), 25, 50,75,df.SellerFeedbackRating.max()]})
X_quartile.head()
```

```python
# As demonstrated below, the less the rating the better the likelihood of winning. 
# From section 1, this is likely a result from isFulfilledbyAmazon. A large amount of winners with IsFulfilledbyAmazon won
# without having any sellerfeedback rating. 
predictions = lm1.predict(X_quartile)
predictions
```

```python
# Plot the quartile range to show the results. 
df.plot(kind='scatter', x="IsWinner", y="SellerFeedbackRating", color="g")

plt.plot(X_quartile, predictions, c='red', linewidth=2)
```

#### 2.1.4. Summary of the trained model.
     - According to the coefficient weight, an increase in SellerFeedbackRating has a decrease in probability in becomming a winner. 
     
     - The t-test checks if there is enough evidence to support or reject the anaylsis. In this case, we have a -41.665 result or more commonly referred to it's absolute value of 41.665 (Source: http://www.csic.cornell.edu/Elrod/t-test/reporting-t-test.html). The negative value simply means that the sample mean was less than the test value. (The negative sign is therefore not all that important when looking at t-values). Instead if the value was null, we would reject the hypothesis, which shouldnt be done in this case.
     
     - Now we look at the p value. If the p value has a confidence rating of above 95% i.e. a p-value lower than 0.05. The p-value is 0 for both intercept and SellerFeedbackRating. As both are less than 0.05, they are are statistically significant.
     
     - The r-squared value evaluates how close the data are to the fitted regression line. "0% indicates that the model explains none of the variability of the response data around its mean" whereas "100% indicates that the model explains all the variability of the response data around its mean"(Source: http://blog.minitab.com/blog/adventures-in-statistics-2/regression-analysis-how-do-i-interpret-r-squared-and-assess-the-goodness-of-fit). The bottom right is the true positives and in our case, the classification correctly predicted the winner in 502 cases. The upper left is a true negative where the classifier correctly predicted that the seller would not win. The upper right is the false positives, which says in 35 cases, it incorrectly predicted the winner, and also the bottom left,(false negatives) in 45 cases, they incorrectly predicted that the seller wouldnt win, but they did. As demonstrated above, the rate of error in our prediction model is very low and therefore there can be a high confidence in our dataset model.
